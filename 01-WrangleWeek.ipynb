```python
import pandas as pd
import boto3
```

# Get Week History

```python
kwargs0 = {'parse_dates': ['Date and Time Auto'], 'date_format': '%m/%d/%y %I:%M %p'}
kwargs1 = {'parse_dates': ['Date and Time Auto'], 'date_format': '%Y-%m-%d %H:%M:%S'}

# df0 = pd.read_csv('raw-2025-06-29.csv', **kwargs0)
# df0 = pd.read_csv('raw-2025-07-06.csv', **kwargs1)
df0 = pd.read_csv('raw-2025-07-13.csv', **kwargs1)

# df0
```

```python
week_history = pd.DataFrame()
week_history['Datetime'] = df0['Date and Time Auto']
week_history['CPM'] = df0['Count Rate (counts/sample) Run #1']
week_history
```

```python
week_name = week_history.iloc[0, 0].strftime('%Y-w%U')
week_name
```

# Export And Upload Week History

```python
file_name = f'history-{week_name}.csv'
file_name
```

```python
week_history.to_csv(file_name, index=False)
```

```python
bucket = boto3.resource('s3').Bucket('hnlrad')
bucket.upload_file(file_name, file_name)
```
