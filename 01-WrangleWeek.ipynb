```python
import pandas
import boto3
```

# Get Week History

```python
raw = 'raw-2026-03-22.csv'

df = pandas.read_csv(raw,
    parse_dates=['Date and Time Auto'],
    date_format='%Y-%m-%d %H:%M:%S')

df = df.rename(columns={
    'Date and Time Auto': 'Datetime',
    'Count Rate (counts/sample) Run #1': 'CPM'})

df
```

# Export And Upload Week History

```python
history = df.iloc[0, 0].strftime('history-%Y-w%U.csv')
history
```

```python
df.to_csv(history, index=False)
```

```python
bucket = boto3.resource('s3').Bucket('hnlrad')
bucket.upload_file(history, history)
```
