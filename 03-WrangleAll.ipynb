```python
import os
import re
import pandas as pd
import boto3
```

# Get All History

```python
pat = re.compile(r'^history-\d\d\d\d-w\d\d\.csv$')
files = [file for file in os.listdir() if pat.match(file)]
files = sorted(files)
files
```

```python
dfs = [pd.read_csv(file) for file in files]
# dfs
```

```python
all_history = pd.concat(dfs, ignore_index=True)
all_history
```

# Export And Upload All History

```python
file_name = f'history-all.csv'
file_name
```

```python
all_history.to_csv(file_name, index=False)
```

```python
bucket = boto3.resource('s3').Bucket('hnlrad')
bucket.upload_file(file_name, file_name)
```
